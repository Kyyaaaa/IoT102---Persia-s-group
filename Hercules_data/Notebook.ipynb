```python
import pandas as pd
df = pd.read_csv("1.csv", header=None, names=["Light(lux)", "Temperature(celsius)", "MQ135(ppm)"])
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tạo index làm trục x (nếu chưa có thời gian)
df["X axis"] = df.index

# Plot từng biểu đồ
columns = ["Light(lux)", "Temperature(celsius)", "MQ135(ppm)"]

for col in columns:
    plt.figure(figsize=(8, 4))
    sns.lineplot(data=df, x="X axis", y=col)
    plt.title(f"{col} over time")
    plt.xlabel("X axis")
    plt.ylabel(col)
    plt.grid(True)
    plt.show()
```
